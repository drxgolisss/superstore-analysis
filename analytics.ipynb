```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

a = pd.read_csv('Sample - Superstore.csv', encoding='ISO-8859-1')

print(a.head())
print(a.info())
print(a.describe())

print(a.isnull().sum())

profit_by_category = a.groupby('Category')['Profit'].sum().sort_values(ascending = False)

plt.figure(figsize=(8, 5))
sns.barplot(x=profit_by_category.index, y=profit_by_category.values)
plt.title('Income by categories')
plt.xlabel('Category')
plt.ylabel('Income')
plt.show()


region_profit = a.groupby('Region')['Profit'].sum()
region_profit.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6), title='Income by regions')
plt.ylabel('')
plt.show()


```
